### 1. Displaying the system of equations (x+3y=4,-2x-9y=5)


```python
x = range(13,21)
line1 = {'x': x, 'y': [(-1/3) * xi + 4/3 for xi in x]}
line2 = {'x': x, 'y': [(-2/9) * xi - 5/9 for xi in x]}

df1 = pd.DataFrame(line1)
df2 = pd.DataFrame(line2)

# Plot the lines
plt.plot(df1['x'], df1['y'], label='x + 3y = 4')
plt.plot(df2['x'], df2['y'], label='-2x -9y = 5')

# Add labels, title, and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

```
